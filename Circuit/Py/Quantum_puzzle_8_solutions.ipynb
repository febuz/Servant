```python
import time
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService
from bitcoin import encode_privkey, privtoaddr


def create_bitcoin_circuit(key, num_qubits):
    """Creates a quantum circuit for private key search."""
    qr = QuantumRegister(num_qubits, 'key')  # Number of qubits for the range
    cr = ClassicalRegister(num_qubits, 'measurement')
    circuit = QuantumCircuit(qr, cr)

    # Encode the key into the qubits
    key_bits = format(key, 'b').zfill(num_qubits)  # Represent the key in binary
    for i, bit in enumerate(key_bits):
        if bit == '1':
            circuit.x(qr[i])

    # Superposition
    circuit.h(qr)

    # Grover-like iterations for amplification
    for _ in range(2):  # Iterations for improved amplification
        # Oracle
        circuit.cz(qr[0], qr[1])  # Simplified oracle for illustration

        # Diffusion
        circuit.h(qr)
        circuit.x(qr)
        circuit.h(qr[-1])
        circuit.mcx(list(range(num_qubits - 1)), qr[-1])  # Multi-controlled NOT
        circuit.h(qr[-1])
        circuit.x(qr)
        circuit.h(qr)

    # Measurement
    circuit.measure(qr, cr)

    return circuit


def is_correct_key(key, target_address):
    """Checks if the given key generates the target address."""
    try:
        private_key_hex = hex(key)[2:].zfill(64)  # Properly format the key
        private_key_wif = encode_privkey(private_key_hex, 'wif_compressed')
        generated_address = privtoaddr(private_key_wif)
        return generated_address == target_address
    except Exception as e:
        print(f"Key validation error: {e}")
        return False


def cleanup_pending_jobs(service, backend_name):
    """Cancels all pending jobs for the given backend."""
    jobs = service.jobs(backend_name=backend_name, limit=10)
    pending_jobs = [job for job in jobs if job.status().name in ['QUEUED', 'RUNNING']]
    print(f"Found {len(pending_jobs)} pending jobs for cleanup.")

    for job in pending_jobs:
        try:
            print(f"Cancelling job {job.job_id()} with status {job.status().name}...")
            job.cancel()
        except Exception as e:
            print(f"Failed to cancel job {job.job_id()}: {e}")

    print("Pending jobs cleanup complete.")


def quantum_search(puzzle_config, backend, service=None):
    """Performs a quantum search for the private key."""
    num_qubits = puzzle_config['num_qubits']
    target_address = puzzle_config['target_address']
    range_start = puzzle_config['range_start']
    range_end = puzzle_config['range_end']

    print(f"Starting quantum search for address: {target_address}")
    print(f"Using backend: {backend.name if service else 'Simulator'}")

    for key in range(range_start, range_end):
        circuit = create_bitcoin_circuit(key, num_qubits)

        try:
            # Transpile the circuit
            transpiled_circuit = transpile(circuit, backend, optimization_level=3)

            if service:
                # Submit the job
                job = backend.run(transpiled_circuit, shots=1024)
                print(f"Submitted job {job.job_id()} for key {key}.")

                # Monitor the job
                while job.status().name in ['QUEUED', 'RUNNING']:
                    print(f"Job {job.job_id()} is {job.status().name}. Waiting for completion...")
                    time.sleep(60)

                if job.status().name == 'DONE':
                    counts = job.result().get_counts()
                    for measured_state, freq in counts.items():
                        measured_int = int(measured_state, 2)

                        if is_correct_key(measured_int, target_address):
                            print(f"\nMatch Found!")
                            print(f"Key: {measured_int} (Hex: {hex(measured_int)})")
                            print(f"Probability: {freq / 1024:.2%}")
                            return measured_int
                else:
                    print(f"Job {job.job_id()} failed with status: {job.status().name}")
            else:
                # Simulate the circuit locally
                job = backend.run(transpiled_circuit, shots=1024)
                results = job.result()
                counts = results.get_counts()

                for measured_state, freq in counts.items():
                    measured_int = int(measured_state, 2)

                    if is_correct_key(measured_int, target_address):
                        print(f"\nMatch Found!")
                        print(f"Key: {measured_int} (Hex: {hex(measured_int)})")
                        print(f"Probability: {freq / 1024:.2%}")
                        return measured_int

            # Display progress
            progress = ((key - range_start) / (range_end - range_start)) * 100
            print(f"Progress: {progress:.1f}% - Checked key: {key} ({hex(key)})")

        except Exception as e:
            print(f"Error processing key {key}: {e}")

    return None


def main():
    """Main function to perform the search."""
    # Puzzle configurations
    puzzles = {
        7: {
            'range_start': 2**6,
            'range_end': 2**7,
            'target_address': "1McVt1vMtCC7yn5b9wgX1833yCcLXzueeC",
            'num_qubits': 7
        },
        8: {
            'range_start': 2**7,
            'range_end': 2**8,
            'target_address': "1M92tSqNmQLYw33fuBvjmeadirh1ysMBxK",
            'num_qubits': 8
        },
        9: {
            'range_start': 2**8,
            'range_end': 2**9,
            'target_address': "1CQFwcjw1dwhtkVWBttNLDtqL7ivBonGPV",
            'num_qubits': 9
        }
    }

    puzzle_number = int(input("Enter puzzle number (7, 8, or 9): "))
    if puzzle_number not in puzzles:
        print("Invalid puzzle number. Please enter 7, 8, or 9.")
        return

    use_hardware = input("Use hardware backend? (y/n): ").strip().lower() == 'y'
    backend = AerSimulator() if not use_hardware else QiskitRuntimeService().backend('ibm_brisbane')
    service = QiskitRuntimeService() if use_hardware else None

    if service:
        print("Cleaning up any pending jobs...")
        cleanup_pending_jobs(service, backend.name)

    print("\nStarting quantum private key search...")
    solution_key = quantum_search(puzzles[puzzle_number], backend, service)

    if solution_key:
        print("\nSOLUTION FOUND!")
        print(f"Private key (decimal): {solution_key}")
        print(f"Private key (hex): {hex(solution_key)[2:].zfill(64)}")
    else:
        print("\nNo solution found in the specified range.")


if __name__ == "__main__":
    main()
```

## QuantumQ quantum level 8 superposition

Hadamard gate

can turn a qubit form a basis state into superposition

|0> --H-- |0> + |1>

|1> --H-- |0> - |1>

*not normalized

The cool thing about a quantum system is superposition. A qubit can be in a state called a superposition.

|0> and |1> are called the basis states, and a qubit can be in a superposition of those two states, which mean, when the qubit is measured, it has some probability to be measured as |0> or |1>. Or what some people would say as being in multiple states at once

A fun interpretation of this phenomenon is that a hadamard gate can split the universe into two versions. One in which the qubit is in the state |0>, and one in which the qubit is in the state |1>

Create a superposition of all 2-bits basis states.

All 2-bits basis states are (|00>, |01>, |10>, |11>)

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService, Session, Sampler
from qiskit.compiler import transpile
from qiskit.visualization import plot_histogram

# Create and show the circuit
qc = QuantumCircuit(2, 2)
qc.h(0)  # Hadamard on first qubit
qc.h(1)  # Hadamard on second qubit
qc.measure_all()

print("Original Circuit:")
print(qc)

# First run on simulator
sim = AerSimulator()
job = sim.run(qc, shots=1000)
result = job.result()
counts = result.get_counts()

print("\nResults from simulator:")
for state, count in counts.items():
    print(f"|{state}> : {count} times ({count/1000:.1%})")

# Now run on real quantum computer
try:
    # Initialize the service
    service = QiskitRuntimeService(channel="ibm_quantum")

    # Get backend
    backend = service.backend('ibm_brisbane')
    print(f"\nUsing backend: {backend.name}")

    # Transpile the circuit for the specific backend
    qc_transpiled = transpile(qc, backend=backend, optimization_level=3)

    print("\nTranspiled Circuit:")
    print(qc_transpiled)

    # Create a session
    with Session(backend=backend) as session:
        # Initialize the sampler within the session
        sampler = Sampler()

        # Run the transpiled circuit
        job = sampler.run([qc_transpiled], shots=1000)
        result = job.result()
        dist = result.quasi_dists[0]

        print("\nResults from quantum computer:")
        for bitstring, prob in dist.items():
            count = int(prob * 1000)
            state = format(bitstring, '02b')
            print(f"|{state}> : {count} times ({prob:.1%})")

        # Add statistical analysis
        print("\nStatistical Analysis:")
        ideal_prob = 0.25
        for bitstring, prob in dist.items():
            state = format(bitstring, '02b')
            deviation = abs(prob - ideal_prob)
            print(f"|{state}> deviation from ideal: {deviation:.3f}")

except Exception as e:
    print(f"\nError running on quantum computer: {str(e)}")
    print("This could be due to backend availability or queue times.")

    # Print available gates on the backend
    try:
        print("\nAvailable gates on backend:")
        print(backend.configuration().basis_gates)
    except:
        pass

# Theoretical explanations
print("\nTheoretical explanation:")
print("Initial state: |00>")
print("After H on first qubit: (|00> + |10>)/√2")
print("After H⊗H: (|00> + |01> + |10> + |11>)/2")
print("\nEach basis state should have approximately 25% probability.")

# Phase information
print("\nPhase Information:")
print("Hadamard transformation:")
print("H|0⟩ = (|0⟩ + |1⟩)/√2")
print("H|1⟩ = (|0⟩ - |1⟩)/√2")

# Detailed evolution
print("\nStep-by-step Evolution:")
print("1. |ψ₀⟩ = |00⟩")
print("2. |ψ₁⟩ = H₁|ψ₀⟩ = (|00⟩ + |10⟩)/√2")
print("3. |ψ₂⟩ = H₂|ψ₁⟩ = (|00⟩ + |01⟩ + |10⟩ + |11⟩)/2")
```
